# Use Python/Folium to marking the earthquack locations
# 使用 Python/Folium 標註地震發生的位置

```python
import folium
import pandas as pd
```

```python
#讀取地震資料
df_earthquake = pd.read_csv(r'D:\earthquake_data\earthquake_2018.csv',encoding='big5')
```

```python
df_earthquack.head()
```

```python
# 顯示地圖/ Show the map
m = folium.Map([23.5, 121], zoom_start=7,tiles='OpenStreetMap')
m
```

```python
# 將座標標註在地圖上/ Let's put some markers on the map
folium.Marker(location=[df_earthquack['緯度'][0],df_earthquack['經度'][0]]).add_to(m)
folium.Marker(location=[df_earthquack['緯度'][2],df_earthquack['經度'][2]]).add_to(m)
folium.Marker(location=[df_earthquack['緯度'][3],df_earthquack['經度'][3]]).add_to(m)
```

試著把今年發生過的地震紀錄全部標註到地圖上<br/>
Now we are going to put all the earthquack records on the map within this year.

```python
# Create a new map object
m_0 = folium.Map([23.5, 121], zoom_start=7,tiles='OpenStreetMap')
```

```python
for index, row in df_earthquack.iterrows():
    folium.Marker(
        location=[row['緯度'],row['經度']]).add_to(m_0)
```

```python
m_0
```

當地圖上有太多標記時導致難以閱讀，利用MarkerCluster將鄰近標註點融合 <br/>
而且指標示出震度規模大於4.5的紀錄<br/>
Since there are too many markers, apply MarkerCluster function to merge the marker which close to each other <br/>
Further more, only the magnitude exceed over 4.5 will be markerd

```python
# Create another new map object
m_1 = folium.Map([23.5, 121], zoom_start=7,tiles='OpenStreetMap')
```

```python
from folium.plugins import MarkerCluster

marker_cluster = MarkerCluster().add_to(m_1)
```

```python
for index, row in df_earthquack.iterrows():
    if row['規模'] >=4.5 :
        information = str(row['地震時間'])+'<br>'+str(row['規模'])+'<br>'+str(row['震央位置'])
        folium.Marker(
            location=[row['緯度'],row['經度']],
            
            popup=information).add_to(marker_cluster)
```
